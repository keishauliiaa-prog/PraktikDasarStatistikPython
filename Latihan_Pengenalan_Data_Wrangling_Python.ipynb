## 🧹 Tahap `Cleaning` Data

### 1️⃣ Membersihkan Data Customers

```python
# Menghapus Duplicate Data
customers_df.drop_duplicates(subset=['customer_id'], inplace=True)  # Hanya menyisakan satu baris per customer_id
print("Jumlah duplikasi setelah cleaning:", customers_df.duplicated().sum())  # Hasil = 0
```

🔹 Keterangan:

* `drop_duplicates()` → menghapus baris identik penuh, memastikan setiap baris unik.

* `.duplicated().sum()` → memastikan jumlah duplikasi = 0.

```python
# Menangani Missing Value
print("\nBaris dengan missing gender:\n", customers_df[customers_df.gender.isna()])  # Menampilkan baris NaN
print("\nDistribusi gender sebelum filling:\n", customers_df.gender.value_counts())  # Melihat distribusi nilai

customers_df.fillna(value="Prefer not to say", inplace=True)  # Mengisi missing gender dengan 'Prefer not to say'
print("\nJumlah missing value setelah filling:\n", customers_df.isna().sum())  # Mengecek kembali, harus 0
```

🔹 Keterangan:

* `fillna(value=...)` → mengganti semua NaN dengan nilai default (`Prefer not to say`).

* `value_counts()` → melihat distribusi kategori sebelum dan sesudah filling.

```python
# Menangani Inaccurate Value / Outlier
print("\nBaris dengan umur maksimum sebelum cleaning:\n", customers_df[customers_df.age == customers_df.age.max()])  # Cek outlier

customers_df.age.replace(customers_df.age.max(), 70, inplace=True)  # Ganti nilai maksimum (misal 700) menjadi 70
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)  # Ganti nilai maksimum baru (misal 500) menjadi 50
```

🔹 Keterangan:

* `replace(old, new, inplace=True)` → mengganti nilai tidak realistis di kolom age.

* Membuat data umur lebih masuk akal untuk analisis.

```python
# Mengecek Statistik Setelah Cleaning
summary_stats = customers_df.describe()  # Ringkasan statistik kolom numerik
print("\nSummary Statistics setelah cleaning:\n", summary_stats)
```

🔹 Keterangan:

* `describe(`) → menampilkan count, mean, std, min, max, median, Q1, Q3.

* Memastikan semua kolom numerik valid dan siap untuk analisis lebih lanjut.

### 2️⃣ Membersihkan Data Order

```python
# Mengubah tipe data tanggal menjadi datetime
datetime_columns = ["order_date", "delivery_date"]   # Daftar kolom yang akan dikonversi

for column in datetime_columns:                      # Loop untuk setiap kolom tanggal
    orders_df[column] = pd.to_datetime(orders_df[column])  # Konversi tipe object → datetime
```

```python
# Menampilkan informasi dataset setelah konversi
orders_df.info()  # Mengecek tipe data kolom, memastikan order_date & delivery_date sudah datetime
```

🔹 Keterangan:

* `pd.to_datetime()` digunakan untuk mengonversi kolom bertipe object (string) menjadi datetime64

* `info()` digunakan untuk memastikan perubahan tipe data berhasil.

* Parameter `inplace=True` berarti perubahan dilakukan langsung pada DataFrame tanpa membuat salinan baru.

# 🔀 Menggabungkan Data: Merge / Join

```python
# Inner Join (hanya mengambil data yang cocok di kedua tabel)
inner_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="inner")  # Hanya customer yang punya order
print("Inner Join (customer yang punya order):")
print(inner_join_df.head(), "\n")

# Left Join (ambil semua data dari kiri + data cocok dari kanan)
left_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="left")  # Semua customer, order bisa NaN
print("Left Join (semua customer, order mungkin kosong):")
print(left_join_df.head(), "\n")

# Right Join (ambil semua data dari kanan + data cocok dari kiri)
right_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="right")  # Semua order, customer mungkin kosong
print("Right Join (semua order, customer mungkin kosong):")
print(right_join_df.head(), "\n")

# Outer Join (ambil semua data dari kedua tabel, gabungkan jika cocok)
outer_join_df = pd.merge(customers_df, orders_df, on="customer_id", how="outer")  # Semua customer dan order
print("Outer Join (semua customer dan order):")
print(outer_join_df.head(), "\n")
```

🔹 Penjelasan Metode Join

1. `on="customer_id"` → kolom yang digunakan untuk menggabungkan kedua DataFrame.

2. how parameter:

* `"inner"` → hanya baris yang cocok di kedua DataFrame.

* `"left"` → semua baris dari DataFrame kiri (customers_df) + data cocok dari kanan.

* `"right"` → semua baris dari DataFrame kanan (orders_df) + data cocok dari kiri.

* `"outer"` → semua baris dari kedua DataFrame, gabungkan jika ada kecocokan.
