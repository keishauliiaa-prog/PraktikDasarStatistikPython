### 🧍‍♀️ Demografi Pelanggan (Customer Demographics)

Tujuan: Mengetahui siapa pelanggan utama kita dan di mana mereka berada untuk mendukung targeted marketing atau ads campaign.

```python
# Distribusi Jenis Kelamin

# Menghitung jumlah masing-masing kategori gender (misal: Male/Female)
gender_summary = customers_df['gender'].value_counts().reset_index()   # Mengembalikan jumlah kemunculan tiap nilai unik dalam kolom 'gender', lalu ubah jadi DataFrame

gender_summary.columns = ['Gender', 'Jumlah']   # Mengganti nama kolom agar lebih mudah dibaca: 'index' → 'Gender', 'gender' → 'Jumlah'

# Menambahkan kolom persentase dari total keseluruhan
gender_summary['Persentase (%)'] = round(      # Membulatkan hasil perhitungan ke dua angka di belakang koma
    (gender_summary['Jumlah'] / gender_summary['Jumlah'].sum()) * 100,  # Rumus: (jumlah per gender / total semua gender) * 100
    2
)

print("Distribusi Jenis Kelamin")  # Menampilkan judul output
print(gender_summary)              # Menampilkan tabel hasil analisis distribusi gender
```

```python
# Statistik Usia

print("Statistik Usia Pelanggan")   # Menampilkan judul bagian analisis

# Menampilkan statistik deskriptif dari kolom 'age'
print(customers_df['age'].describe())  # Fungsi describe() memberikan ringkasan statistik:
                                       # - count   : jumlah data usia yang valid (tidak null)
                                       # - mean    : rata-rata usia
                                       # - std     : standar deviasi (tingkat penyebaran usia)
                                       # - min     : usia minimum
                                       # - 25%     : kuartil pertama (batas bawah 25% data)
                                       # - 50%     : median (titik tengah data)
                                       # - 75%     : kuartil ketiga (batas atas 75% data)
                                       # - max     : usia maksimum
```

```python
# Distribusi Kelompok Usia

# Menentukan batas (range) usia untuk pengelompokan
bins = [0, 20, 30, 40, 50, 60, 70, 100]      # Rentang interval usia (batas bawah dan atas setiap kelompok)
labels = ['<20', '21-30', '31-40', '41-50',  # Label yang merepresentasikan setiap rentang usia
          '51-60', '61-70', '70+']

# Membuat kolom baru 'age_group' berdasarkan pembagian usia menggunakan pd.cut()
customers_df['age_group'] = pd.cut(
    customers_df['age'],                     # Kolom usia yang akan dikelompokkan
    bins=bins,                               # Daftar rentang usia
    labels=labels,                           # Label untuk setiap kelompok
    right=False                              # Interval kiri-tertutup, kanan-terbuka (misal: 20 masuk ke 20–30, bukan <20)
)

# Menghitung jumlah pelanggan di setiap kelompok usia
age_group_summary = customers_df['age_group'].value_counts()  # Hitung jumlah tiap kategori 'age_group'
age_group_summary = age_group_summary.sort_index()            # Urutkan sesuai urutan label (bukan abjad)
age_group_summary = age_group_summary.reset_index()           # Ubah jadi DataFrame agar bisa diolah lebih lanjut

age_group_summary.columns = ['Kelompok Usia', 'Jumlah']       # Ganti nama kolom agar lebih deskriptif

# Hitung persentase setiap kelompok terhadap total
age_group_summary['Persentase (%)'] = round(
    (age_group_summary['Jumlah'] / age_group_summary['Jumlah'].sum()) * 100, 2
)

print("Distribusi Berdasarkan Kelompok Usia")  # Tampilkan judul
print(age_group_summary)                       # Tampilkan tabel hasil distribusi usia
```

```python
# Distribusi Negara

# Menghitung jumlah pelanggan berdasarkan negara
country_summary = customers_df['country'].value_counts().reset_index()   # Hitung jumlah data untuk tiap negara, lalu ubah hasil Series jadi DataFrame

country_summary.columns = ['Negara', 'Jumlah']   # Ganti nama kolom agar lebih informatif

# Hitung persentase tiap negara terhadap total pelanggan
country_summary['Persentase (%)'] = round(
    (country_summary['Jumlah'] / country_summary['Jumlah'].sum()) * 100,  # Rumus: (jumlah per negara / total pelanggan) * 100
    2                                                                     # Hasil dibulatkan ke 2 angka desimal
)

print("Distribusi Berdasarkan Negara")   # Menampilkan judul output
print(country_summary)                   # Menampilkan tabel hasil distribusi pelanggan berdasarkan negara
```

```python
# State Teratas

# Menghitung jumlah pelanggan berdasarkan 'state' dan menampilkan hanya 10 teratas
state_summary = customers_df['state'].value_counts()   # Hitung jumlah data untuk tiap state (provinsi/wilayah)
state_summary = state_summary.head(10)                 # Ambil hanya 10 state dengan jumlah pelanggan terbanyak
state_summary = state_summary.reset_index()            # Ubah hasil Series menjadi DataFrame agar mudah diolah

state_summary.columns = ['State', 'Jumlah']            # Ganti nama kolom agar lebih mudah dibaca

print("10 State dengan Jumlah Pelanggan Terbanyak")    # Tampilkan judul analisis
print(state_summary)                                   # Tampilkan tabel hasil
```

```python
# Kota Teratas

# Menghitung jumlah pelanggan berdasarkan kota dan mengambil 10 teratas
city_summary = customers_df['city'].value_counts()     # Hitung jumlah kemunculan tiap kota (berapa banyak pelanggan di tiap kota)
city_summary = city_summary.head(10)                   # Ambil hanya 10 kota dengan jumlah pelanggan terbanyak
city_summary = city_summary.reset_index()              # Ubah hasil Series jadi DataFrame agar mudah diolah dan diberi nama kolom

city_summary.columns = ['Kota', 'Jumlah']              # Ganti nama kolom agar lebih mudah dibaca

print("10 Kota dengan Jumlah Pelanggan Terbanyak")     # Tampilkan judul hasil analisis
print(city_summary)                                    # Tampilkan tabel hasil top 10 kota
```

* `pd.cut()` → digunakan untuk mengelompokkan usia ke dalam kategori umur tertentu.

### 🛍️ Aktivitas Order Pelanggan
Tujuan: Mengukur seberapa banyak pelanggan yang benar-benar aktif melakukan pembelian dan seberapa sering mereka bertransaksi.

```python
import pandas as pd   # Mengimpor library pandas untuk manipulasi data

# Menggabungkan Data Order dengan Data Customer
orders_geo = pd.merge(                # Fungsi merge() digunakan untuk menggabungkan dua DataFrame
    orders_df,                        # DataFrame pertama → berisi data pesanan (orders)
    customers_df,                     # DataFrame kedua → berisi data pelanggan (customers)
    on='customer_id',                 # Kolom kunci penghubung di kedua tabel (foreign key)
    how='left'                        # Jenis join: LEFT JOIN → semua data dari orders_df dipertahankan,
                                      # dan data customers_df ditambahkan jika ada customer_id yang cocok
)
```

```python
# Hitung jumlah order per pelanggan
customer_activity = orders_df.groupby('customer_id').agg(   # Mengelompokkan data order berdasarkan customer_id
    jumlah_order=('order_id', 'count'),                     # Hitung banyaknya order yang dilakukan setiap pelanggan
    total_pembayaran=('payment', 'sum')                     # Jumlahkan total pembayaran dari setiap pelanggan
).reset_index()                                             # Kembalikan hasil ke bentuk DataFrame biasa

# Gabungkan hasil agregasi dengan data pelanggan
customer_activity = pd.merge(                               # Menggabungkan data aktivitas dengan informasi pelanggan
    customer_activity,                                       # Data hasil agregasi (jumlah order dan total pembayaran)
    customers_df[['customer_id', 'gender', 'age', 'country']], # Kolom penting dari data pelanggan
    on='customer_id',                                       # Kolom penghubung antara kedua DataFrame
    how='left'                                              # LEFT JOIN → semua pelanggan yang pernah order tetap ada
)

# Statistik umum aktivitas pelanggan
total_customer = customers_df['customer_id'].nunique()      # Hitung total pelanggan unik yang terdaftar
customer_order = customer_activity['customer_id'].nunique() # Hitung jumlah pelanggan yang pernah melakukan order
inactive_customer = total_customer - customer_order         # Hitung pelanggan yang belum pernah melakukan order

print("Statistik Aktivitas Pelanggan")                      # Tampilkan judul analisis
print(f"Total Pelanggan Terdaftar       : {total_customer}")  # Jumlah pelanggan keseluruhan
print(f"Pelanggan yang Pernah Order     : {customer_order}")  # Pelanggan aktif (pernah order)
print(f"Pelanggan Tidak Pernah Order    : {inactive_customer}") # Pelanggan pasif (belum pernah order)
print(f"Persentase Pelanggan Aktif (%)  : {round((customer_order / total_customer) * 100, 2)}%") # Rasio pelanggan aktif
```

```python
# Kelompokkan pelanggan berdasarkan aktivitasnya

# Fungsi untuk mengelompokkan pelanggan berdasarkan jumlah transaksi
def kategori_order(jumlah):                    # Parameter 'jumlah' = jumlah_order tiap pelanggan
    if jumlah == 1:                            # Jika hanya 1 kali transaksi
        return '1x Transaksi'
    elif 2 <= jumlah <= 3:                     # Jika 2–3 kali transaksi
        return '2-3x Transaksi'
    elif 4 <= jumlah <= 5:                     # Jika 4–5 kali transaksi
        return '4-5x Transaksi'
    else:                                      # Jika lebih dari 5 kali transaksi
        return '>5x Transaksi'

# Terapkan fungsi kategori_order pada kolom 'jumlah_order'
customer_activity['kategori_aktivitas'] = customer_activity['jumlah_order'].apply(kategori_order)
# → apply() menerapkan fungsi custom ke setiap baris kolom tertentu

# Hitung jumlah pelanggan per kategori aktivitas
aktivitas_summary = customer_activity['kategori_aktivitas'].value_counts().reset_index()
# value_counts() menghitung jumlah pelanggan di tiap kategori
# reset_index() mengubah hasilnya menjadi DataFrame agar mudah diolah

aktivitas_summary.columns = ['Kategori Aktivitas', 'Jumlah Pelanggan']  # Ganti nama kolom agar lebih jelas

# Hitung persentase setiap kategori terhadap total pelanggan
aktivitas_summary['Persentase (%)'] = round(
    (aktivitas_summary['Jumlah Pelanggan'] / aktivitas_summary['Jumlah Pelanggan'].sum()) * 100, 2
)

print("Distribusi Pelanggan Berdasarkan Frekuensi Transaksi")  # Judul hasil analisis
print(aktivitas_summary)                                       # Tampilkan tabel hasil
```

```python
# Top 10 Pelanggan dengan Total Pembayaran Tertinggi

# Urutkan pelanggan berdasarkan total pembayaran dari terbesar ke terkecil
top_customers = customer_activity.sort_values(
    by='total_pembayaran',       # Kolom yang dijadikan patokan urutan
    ascending=False              # Descending → pembayaran terbesar di atas
).head(10)                       # Ambil 10 pelanggan teratas

print("10 Pelanggan dengan Total Pembayaran Tertinggi")  # Judul output

# Tampilkan kolom penting saja agar lebih ringkas
print(top_customers[['customer_id', 'jumlah_order', 'total_pembayaran', 'country']])
```

🔹 Keterangan:

* `pd.merge()` → digunakan untuk menggabungkan data pelanggan (`customers_df`) dengan data order (`orders_df`) berdasarkan `customer_id`.

* `groupby('customer_id').agg()` → menghitung jumlah order dan total pembayaran per pelanggan.

* Fungsi `kategori_order()` → mengelompokkan pelanggan berdasarkan frekuensi transaksi (`1x`, `2–3x`, `4–5x`, `>5x`).

* `value_counts()` → menghitung distribusi jumlah pelanggan di tiap kategori aktivitas transaksi.

🔹 Insight Penting:

* Tingkat Aktivitas Pelanggan:
Dari total 1000 pelanggan, sebanyak 617 pelanggan (61,7%) pernah melakukan transaksi, sementara 383 pelanggan (38,3%) belum pernah berbelanja.
> Lebih dari separuh pelanggan aktif, namun masih ada potensi besar untuk mengaktivasi pelanggan pasif.

* Frekuensi Transaksi:
Sebagian besar pelanggan hanya bertransaksi 1 kali (`57,37%`), dan `39,38%` melakukan pembelian 2–3 kali.
>  Menunjukkan loyalitas pelanggan masih rendah, perlu strategi retensi seperti voucher, loyalty point, atau personalisasi promo.

* Pelanggan Premium (High Spender):
Pelanggan dengan ID 664, 566, dan 282 mencatat total pembayaran tertinggi, masing-masing lebih dari `200 ribu AUD` dengan frekuensi `4–6 kali` transaksi.
> Segmen pelanggan ini bisa dijadikan target program loyalitas atau VIP karena kontribusinya besar terhadap total revenue.

* Distribusi Negara:
Seluruh pelanggan berasal dari `Australia`, sehingga pengelompokan aktivitas dapat difokuskan pada level demografi seperti usia dan wilayah bagian (state).

### 🌏 Demografi Persebaran Data Order

Tujuan: Mengetahui lokasi geografis yang memiliki aktivitas pembelian paling tinggi.

```python
# Jumlah Order per Negara

# Mengelompokkan data order berdasarkan negara dan menghitung jumlah order & total pembayaran
country_orders = (
    orders_geo.groupby('country').agg(
        jumlah_order=('order_id', 'count'),    # Hitung jumlah order untuk tiap negara
        total_pembayaran=('payment', 'sum')   # Jumlahkan total pembayaran untuk tiap negara
    )
    .reset_index()                             # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='jumlah_order', ascending=False)  # Urutkan berdasarkan jumlah order terbesar
)

print("Persebaran Order Berdasarkan Negara")   # Judul output
print(country_orders)                          # Tampilkan tabel hasil
```

```python
# 10 State dengan Jumlah Order Tertinggi

# Mengelompokkan data order berdasarkan state dan menghitung jumlah order & total pembayaran
state_orders = (
    orders_geo.groupby('state').agg(
        jumlah_order=('order_id', 'count'),      # Hitung jumlah order untuk tiap state
        total_pembayaran=('payment', 'sum')     # Jumlahkan total pembayaran untuk tiap state
    )
    .reset_index()                               # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='jumlah_order', ascending=False)  # Urutkan berdasarkan jumlah order terbesar
    .head(10)                                    # Ambil 10 state teratas
)

print("10 State dengan Jumlah Order Tertinggi")  # Judul output
print(state_orders)                               # Tampilkan tabel hasil
```

```python
# 10 Kota dengan Jumlah Order Tertinggi

# Mengelompokkan data order berdasarkan kota dan menghitung jumlah order & total pembayaran
city_orders = (
    orders_geo.groupby('city').agg(
        jumlah_order=('order_id', 'count'),      # Hitung jumlah order untuk tiap kota
        total_pembayaran=('payment', 'sum')     # Jumlahkan total pembayaran untuk tiap kota
    )
    .reset_index()                               # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='jumlah_order', ascending=False)  # Urutkan berdasarkan jumlah order terbesar
    .head(10)                                    # Ambil 10 kota teratas
)

print("10 Kota dengan Jumlah Order Tertinggi")   # Judul output
print(city_orders)                               # Tampilkan tabel hasil
```

🔹 Keterangan:

* `pd.merge()` → digunakan untuk menggabungkan data order (`orders_df`) dengan data pelanggan (`customers_df`) agar setiap order memiliki informasi lokasi.

* `groupby() + agg()` → menghitung jumlah order (`jumlah_order`) dan total pembayaran (total_pembayaran) berdasarkan wilayah seperti negara, state, dan kota.

* `sort_values()` → mengurutkan hasil berdasarkan jumlah order tertinggi agar mudah mengidentifikasi wilayah dengan aktivitas tertinggi.

* Karena seluruh pelanggan berasal dari Australia, analisis fokus pada state dan city sebagai level geografis pembeda.

🔹 Insight Penting:

* Pusat Aktivitas Utama:
Seluruh transaksi berasal dari Australia dengan total `1.000` order dan nilai pembayaran mencapai `33,9 juta AUD`.
> Menunjukkan pasar masih terfokus di satu negara, namun potensial untuk ekspansi internasional.

* Top 3 State dengan Aktivitas Order Tertinggi: `South Australia` – 148 order (AUD 4,87 juta), `Queensland` – 139 order (AUD 4,90 juta), `Western Australia` – 130 order (AUD 4,25 juta)
> Ketiga wilayah ini menyumbang volume dan pendapatan terbesar — cocok dijadikan prioritas kampanye regional.

* Performa Kota:
Kota `Jordanside` dan `New Ava` mencatat jumlah order tertinggi (`6 order`), sedangkan East Max mencatat nilai transaksi tertinggi (`224.843 AUD`).

* Pola Geografis yang Stabil:

### 👕 Demografi Persebaran Data Produk

Tujuan: Menganalisis performa produk berdasarkan tipe, warna, ukuran, dan harga untuk mengidentifikasi produk paling populer.

```python
import pandas as pd   # Mengimpor library pandas untuk manipulasi data

# Gabungkan data produk dengan data penjualan
product_sales = pd.merge(
    sales_df,                   # DataFrame berisi data penjualan
    products_df,                # DataFrame berisi informasi produk
    on='product_id',            # Kolom kunci penghubung di kedua DataFrame
    how='left'                  # LEFT JOIN → semua data penjualan tetap ada, info produk ditambahkan jika cocok
)
```

```python
# Pastikan nama kolom 'quantity' benar
# Kadang merge menyebabkan duplikasi kolom (quantity_x dan quantity_y)
if 'quantity_x' in product_sales.columns:
    product_sales.rename(columns={'quantity_x': 'quantity'}, inplace=True)  # Gunakan kolom dari sales_df
elif 'quantity_y' in product_sales.columns:
    product_sales.rename(columns={'quantity_y': 'quantity'}, inplace=True)  # Gunakan kolom dari products_df jika ada

# Hitung total penjualan per produk
product_summary = product_sales.groupby(
    ['product_id', 'product_type', 'product_name', 'size', 'colour']  # Kelompokkan berdasarkan atribut produk
).agg(
    total_terjual=('quantity', 'sum'),            # Total jumlah unit terjual
    total_pendapatan=('total_price', 'sum'),      # Total pendapatan dari penjualan produk
    harga_rata_rata=('price_per_unit', 'mean')    # Rata-rata harga per unit (bisa untuk analisis margin/harga)
).reset_index()                                   # Kembalikan hasil groupby menjadi DataFrame biasa
```

```python
# Analisis berdasarkan tipe produk

# Mengelompokkan produk berdasarkan 'product_type' dan menghitung performa
type_summary = (
    product_summary.groupby('product_type').agg(
        total_item_terjual=('total_terjual', 'sum'),   # Jumlah total unit terjual per tipe produk
        total_pendapatan=('total_pendapatan', 'sum'),  # Total pendapatan dari tiap tipe produk
        harga_rata_rata=('harga_rata_rata', 'mean')    # Rata-rata harga per unit per tipe produk
    )
    .reset_index()                                      # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='total_item_terjual', ascending=False)  # Urutkan dari tipe produk yang terjual terbanyak
)

print("Performa Berdasarkan Tipe Produk")  # Judul output
print(type_summary)                        # Tampilkan tabel hasil analisis
```

```python
# Analisis berdasarkan warna produk

# Mengelompokkan produk berdasarkan 'colour' dan menghitung performa
colour_summary = (
    product_summary.groupby('colour').agg(
        total_item_terjual=('total_terjual', 'sum'),   # Jumlah total unit terjual per warna
        total_pendapatan=('total_pendapatan', 'sum')   # Total pendapatan dari tiap warna
    )
    .reset_index()                                      # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='total_item_terjual', ascending=False)  # Urutkan berdasarkan jumlah item terjual terbanyak
)

print("Performa Berdasarkan Warna Produk")  # Judul output
print(colour_summary)                       # Tampilkan tabel hasil analisis
```

```python
# Analisis berdasarkan ukuran produk

# Mengelompokkan produk berdasarkan 'size' dan menghitung performa
size_summary = (
    product_summary.groupby('size').agg(
        total_item_terjual=('total_terjual', 'sum'),   # Jumlah total unit terjual per ukuran
        total_pendapatan=('total_pendapatan', 'sum')   # Total pendapatan dari tiap ukuran
    )
    .reset_index()                                      # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='total_item_terjual', ascending=False)  # Urutkan berdasarkan jumlah item terjual terbanyak
)

print("Performa Berdasarkan Ukuran Produk")  # Judul output
print(size_summary)                          # Tampilkan tabel hasil analisis
```

```python
# 10 Produk Terpopuler

# Urutkan produk berdasarkan total unit terjual dari terbesar ke terkecil
top_products = product_summary.sort_values(
    by='total_terjual',    # Kolom patokan urutan → jumlah unit terjual
    ascending=False         # Descending → produk terjual terbanyak di atas
).head(10)                  # Ambil 10 produk teratas

print("10 Produk Paling Populer (Berdasarkan Jumlah Penjualan)")  # Judul output

# Tampilkan kolom penting saja agar lebih ringkas
print(top_products[['product_name', 'product_type', 'colour', 'size', 'total_terjual', 'total_pendapatan']])
```

### 📈 Tren Penjualan Setiap Bulan

Tujuan: Menganalisis pola dan musim penjualan untuk mendukung perencanaan stok dan strategi kampanye promosi.

```python
import pandas as pd   # Mengimpor library pandas untuk manipulasi data

# Pastikan kolom 'order_date' dalam format datetime agar bisa diolah sebagai tanggal
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])

# Gabungkan data penjualan dengan data order untuk mendapatkan tanggal tiap transaksi
sales_orders = pd.merge(
    sales_df,                  # DataFrame berisi data penjualan
    orders_df,                 # DataFrame berisi informasi order (termasuk tanggal)
    on='order_id',             # Kolom penghubung antar DataFrame
    how='left'                 # LEFT JOIN → semua data penjualan tetap ada
)

# Tambahkan kolom bulan dan tahun dari tanggal order
sales_orders['bulan'] = sales_orders['order_date'].dt.month        # Ambil nomor bulan (1–12)
sales_orders['tahun'] = sales_orders['order_date'].dt.year         # Ambil tahun dari tanggal
sales_orders['periode'] = sales_orders['order_date'].dt.to_period('M').astype(str)
# 'periode' = format YYYY-MM → mempermudah agregasi bulanan
```

```python
# Hitung total pendapatan dan jumlah order per bulan

monthly_trend = (
    sales_orders.groupby('periode').agg(
        total_pendapatan=('total_price', 'sum'),   # Total pendapatan tiap bulan
        total_order=('order_id', 'nunique')        # Jumlah order unik tiap bulan
    )
    .reset_index()                                  # Kembalikan hasil groupby menjadi DataFrame biasa
    .sort_values(by='periode')                      # Urutkan berdasarkan periode (YYYY-MM)
)

print("Tren Penjualan Setiap Bulan")               # Judul output
print(monthly_trend)                               # Tampilkan tabel hasil
```

```python
# Identifikasi bulan dengan performa terbaik dan terendah

# Bulan dengan total pendapatan tertinggi
best_month = monthly_trend.loc[monthly_trend['total_pendapatan'].idxmax()]
# idxmax() → mengembalikan index baris dengan nilai maksimum di kolom 'total_pendapatan'

# Bulan dengan total pendapatan terendah
worst_month = monthly_trend.loc[monthly_trend['total_pendapatan'].idxmin()]
# idxmin() → mengembalikan index baris dengan nilai minimum di kolom 'total_pendapatan'

print("Bulan dengan Penjualan Tertinggi")
print(best_month)                           # Tampilkan bulan dengan pendapatan tertinggi
print("\n")

print("Bulan dengan Penjualan Terendah")
print(worst_month)                          # Tampilkan bulan dengan pendapatan terendah
```

🔹 Keterangan:

* `.loc` → indexer yang digunakan untuk mengakses baris dan kolom DataFrame

* Data tren penjualan diperoleh dengan menggabungkan dataset orders_df dan sales_df, lalu dikelompokkan berdasarkan periode bulan (`periode`).

* Kolom `order_date` telah dikonversi ke format datetime untuk memudahkan ekstraksi bulan dan tahun.

* `total_pendapatan` → total nilai penjualan per bulan.

* `total_order` → jumlah pesanan unik per bulan.

* Penjualan tertinggi terjadi pada Maret 2021, dengan total pendapatan 131.364 dan 117 order, menunjukkan adanya momentum positif (kemungkinan kampanye atau tren musiman).

* Penjualan terendah tercatat pada Oktober 2021, hanya 84.266 pendapatan dengan 80 order, menandakan perlunya strategi promosi tambahan di periode akhir tahun.
