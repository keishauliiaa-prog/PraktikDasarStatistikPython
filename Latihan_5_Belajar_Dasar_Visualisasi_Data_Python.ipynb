# 🎨 Visualisasi Data pada Python

## 1️⃣ Bar Chart – Penjualan Tiap Minggu per Bulan

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    'Bulan': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni'],
    'Minggu 1': [500, 600, 550, 500, 650, 700],
    'Minggu 2': [450, 580, 530, 520, 640, 710],
    'Minggu 3': [480, 610, 560, 530, 630, 720],
    'Minggu 4': [510, 620, 570, 540, 620, 730]
}

df = pd.DataFrame(data)                    # Buat tabel DataFrame

x = np.arange(len(df['Bulan']))            # Posisi untuk tiap bulan di sumbu X
width = 0.2                                # Lebar batang

plt.figure(figsize=(10,6))                 # Atur ukuran grafik
plt.bar(x - 1.5*width, df['Minggu 1'], width, label='Minggu 1')  # Batang Minggu 1
plt.bar(x - 0.5*width, df['Minggu 2'], width, label='Minggu 2')  # Batang Minggu 2
plt.bar(x + 0.5*width, df['Minggu 3'], width, label='Minggu 3')  # Batang Minggu 3
plt.bar(x + 1.5*width, df['Minggu 4'], width, label='Minggu 4')  # Batang Minggu 4

plt.title("Perbandingan Penjualan Tiap Minggu per Bulan")  # Judul grafik
plt.xlabel("Bulan")                                        # Label sumbu X
plt.ylabel("Penjualan (unit)")                             # Label sumbu Y
plt.xticks(x, df['Bulan'])                                 # Nama bulan pada sumbu X
plt.legend()                                               # Tampilkan legenda
plt.grid(linestyle='--', alpha=0.6)                        # Garis bantu horizontal

plt.show()
```

## 2️⃣ Line Chart – Tren Penjualan Bulanan

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Bulan': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni',
              'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Penjualan': [150, 180, 200, 210, 240, 250, 300, 320, 340, 330, 290, 310]
}

df = pd.DataFrame(data)                    # Buat DataFrame

plt.figure(figsize=(10, 6))                # Ukuran grafik
plt.plot(df['Bulan'], df['Penjualan'],     # Data X dan Y
         marker='o', color='royalblue', linewidth=2, label='Total Penjualan')  # Bentuk dan warna garis

plt.title('Tren Penjualan Bulanan Tahun 2025', fontsize=12)  # Judul grafik
plt.xlabel('Bulan', fontsize=12)                             # Label sumbu X
plt.ylabel('Jumlah Penjualan', fontsize=12)                  # Label sumbu Y
plt.grid(True, linestyle='--', alpha=0.6)                    # Grid
plt.legend()                                                 # Tampilkan legenda
plt.tight_layout()                                           # Layout rapi
plt.show()
```

## 3️⃣ Stacked Area Chart – Akumulasi Penggunaan Sistem Operasi

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Bulan': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni',
              'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Windows': [40, 42, 45, 48, 50, 53, 55, 57, 58, 60, 62, 65],
    'macOS': [20, 21, 23, 25, 27, 28, 30, 32, 33, 35, 36, 38],
    'Linux': [15, 17, 18, 19, 20, 22, 23, 24, 26, 27, 28, 30]
}

df = pd.DataFrame(data)  # Buat DataFrame

plt.figure(figsize=(10,6))
plt.stackplot(df['Bulan'], df['Windows'], df['macOS'], df['Linux'],
              labels=['Windows', 'macOS', 'Linux'], colors=['blue','orange','green'])  # Stacked area

plt.xlabel('Bulan')
plt.ylabel('Penggunaan (Dalam Ribuan)')
plt.title('Akumulasi Penggunaan Sistem Operasi Bulanan')
plt.legend(loc='upper left')    # Tampilkan legenda
plt.xticks(rotation=45)         # Rotasi label bulan
plt.tight_layout()
plt.show()
```

## 4️⃣ Pie Chart – Proporsi Hobi Siswa

```python
import pandas as pd
import matplotlib.pyplot as plt

# 🔹 Data
data = {
    'Hobi': ['Membaca', 'Bermain Musik', 'Olahraga', 'Melukis', 'Lainnya'],
    'Persentase': [25, 20, 30, 15, 10]
}

df = pd.DataFrame(data)  # Buat DataFrame

plt.figure(figsize=(6,6))
plt.pie(df['Persentase'], labels=df['Hobi'], autopct='%1.1f%%', startangle=90,
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])  # Pie chart

plt.axis('equal')          # Lingkaran proporsional
plt.title('Proporsi Hobi Siswa')
plt.show()
```

## 5️⃣ Scatter Plot – Hubungan Pendapatan dan Kebahagiaan

```python
import pandas as pd
import matplotlib.pyplot as plt

# 🔹 Data
data = {
    'Income': [3, 4, 5, 6, 7, 8, 9, 10, 12, 15],
    'Happiness': [4, 5, 6, 7, 8, 8.5, 9, 9.5, 10, 10]
}

df = pd.DataFrame(data)  # Buat DataFrame

plt.figure(figsize=(8,6))

# Pendapatan <= 8 juta
subset1 = df[df['Income'] <= 8]
plt.scatter(subset1['Income'], subset1['Happiness'], color='red', marker='x', label='Pendapatan <= 8 juta')

# Pendapatan > 8 juta
subset2 = df[df['Income'] > 8]
plt.scatter(subset2['Income'], subset2['Happiness'], color='blue', marker='o', label='Pendapatan > 8 juta')

plt.xlabel('Pendapatan (Juta/Bulan)')
plt.ylabel('Kebahagiaan (Skala 1-10)')
plt.title('Hubungan Antara Pendapatan dan Kebahagiaan')
plt.legend()
plt.show()
```

| Fungsi                          | Keterangan                                                                           |
| ------------------------------- | ------------------------------------------------------------------------------------ |
| `pd.DataFrame(data)`            | Membuat DataFrame dari dictionary, kolom = key, nilai = list                         |
| `plt.bar()`                     | Membuat bar chart; `x` = posisi, `height` = tinggi batang                            |
| `plt.plot()`                    | Membuat line chart; `marker` = simbol titik, `linewidth` = ketebalan garis           |
| `plt.stackplot()`               | Membuat stacked area chart untuk akumulasi data                                      |
| `plt.pie()`                     | Membuat diagram lingkaran; `autopct` = format persen, `labels` = label tiap potongan |
| `plt.scatter()`                 | Membuat scatter plot; `marker` = simbol titik, `color` = warna                       |
| `plt.xlabel()` / `plt.ylabel()` | Memberi label pada sumbu X dan Y                                                     |
| `plt.title()`                   | Memberi judul grafik                                                                 |
| `plt.legend()`                  | Menampilkan legenda untuk keterangan                                                 |
| `plt.grid()`                    | Menampilkan garis bantu pada grafik                                                  |
| `plt.tight_layout()`            | Mengatur layout agar elemen grafik tidak terpotong                                   |
| `plt.show()`                    | Menampilkan grafik                                                                   |
